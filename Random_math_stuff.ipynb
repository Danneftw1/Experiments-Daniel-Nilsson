# Sine & Cosine Waves

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.show()
```

# 3D Parametric Cruve - Helix

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 10 * np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Parametric Curve - Helix')
plt.show()
```

# 2D Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_2d(x, y, mu, sigma):
    return np.exp(-((x - mu[0])**2 + (y - mu[1])**2) / (2 * sigma**2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

mu = [0, 0]
sigma = 1
Z = gaussian_2d(X, Y, mu, sigma)

plt.contourf(X, Y, Z, 20, cmap='plasma')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Gaussian Distribution')
plt.colorbar()
plt.show()
```

# Lissajous cruves

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 1000)
a_x, a_y = 5, 4
delta = np.pi / 2

x = np.sin(a_x * t + delta)
y = np.sin(a_y * t)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lissajous Curve')
plt.show()
```

# Bifurcation diagram of the logistic map

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

iterations = 1000
discard = 200
r_values = np.linspace(2.5, 4.0, 1000)
x_values = []

for r in r_values:
    x = np.random.rand()
    for _ in range(iterations):
        x = logistic_map(r, x)
        if _ > discard:
            x_values.append((r, x))

r, x = zip(*x_values)
plt.scatter(r, x, s=0.5, alpha=0.1)
plt.xlabel('r')
plt.ylabel('x')
plt.title('Bifurcation Diagram of the Logistic Map')
plt.show()
```

# 3D Surface (Saddle point)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface - Saddle Point')
plt.show()
```

# Lorenz attractor

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import solve_ivp

def lorenz_system(t, xyz, sigma, beta, rho):
    x, y, z = xyz
    return [
        sigma * (y - x),
        x * (rho - z) - y,
        x * y - beta * z
    ]

t_span = (0, 50)
initial_state = [1, 1, 1]
sigma, beta, rho = 10, 8/3, 28

sol = solve_ivp(
    lorenz_system, t_span, initial_state, args=(sigma, beta, rho),
    dense_output=True, max_step=0.01
)

t = np.linspace(t_span[0], t_span[1], 10000)
xyz = sol.sol(t)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xyz[0], xyz[1], xyz[2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Lorenz Attractor')
plt.show()
```
