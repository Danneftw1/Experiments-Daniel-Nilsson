# Sine & Cosine Waves

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.show()
```

# 3D Parametric Cruve - Helix

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 10 * np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Parametric Curve - Helix')
plt.show()
```

# 2D Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_2d(x, y, mu, sigma):
    return np.exp(-((x - mu[0])**2 + (y - mu[1])**2) / (2 * sigma**2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

mu = [0, 0]
sigma = 1
Z = gaussian_2d(X, Y, mu, sigma)

plt.contourf(X, Y, Z, 20, cmap='plasma')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Gaussian Distribution')
plt.colorbar()
plt.show()
```

# Lissajous cruves

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 1000)
a_x, a_y = 5, 4
delta = np.pi / 2

x = np.sin(a_x * t + delta)
y = np.sin(a_y * t)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lissajous Curve')
plt.show()
```
